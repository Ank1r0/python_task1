10. Посчитайте среднее время работы (hours-per-week) зарабатывающих мало и много (salary) для каждой страны (native-country).

```python
rich_people = df[df['salary'] == '>50K']
poor_people = df[df['salary'] == '<=50K']
workingTimeAvgPoor = poor_people.groupby(['native-country'])[['hours-per-week']].mean()
workingTimeAvgRich = rich_people.groupby(['native-country'])[['hours-per-week']].mean()

poor_col = workingTimeAvgPoor.rename(columns={'hours-per-week': 'Avg Hours <=50K'})
rich_col = workingTimeAvgRich.rename(columns={'hours-per-week': 'Avg Hours >50K'})

# Join them together
comparison_df = pd.concat([poor_col, rich_col], axis=1).sort_values(by='Avg Hours <=50K',ascending=False)

print(comparison_df)
```

11.Сгруппируйте людей по возрастным группам young, adult, retiree, где:
young соответствует 16-35 лет
adult - 35-70 лет
retiree - 70-100 лет
Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup

```python
df_analysis = df.copy()

bins = [0, 16, 35, 70, 100]
labels = ['minor', 'young', 'adult', 'retiree']

df_analysis['AgeGroup'] = pd.cut(df_analysis['age'], bins=bins, labels=labels)

print(df_analysis[['age', 'AgeGroup']].head())
```

12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)

```python
#12
rich_people = df_analysis[df_analysis['salary'] == '>50K'].groupby(['AgeGroup']).size()

print(rich_people)


#13
percentages = df_analysis.groupby('AgeGroup')['salary'].apply(lambda x: (x == '>50K').mean())

# Find the name of the group with the maximum value
most_often_group = percentages.idxmax()

print(f"\nThe group that earns >50K most often is: {most_often_group}")
print(f"Success rate: {percentages.max() * 100:.1f}%")
```

14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)

```python
import pandas as pd

# 1. Group by occupation and show initial counts
occupation_counts = df.groupby('occupation').size()
print("Initial counts per occupation:")
print(occupation_counts)
print("-" * 30)

# 2. Define the filter function
def filter_func(group):
    # Check if average age is <= 40
    avg_age_ok = group['age'].mean() <= 40
    # Check if EVERY worker in this group works more than 5 hours
    min_hours_ok = (group['hours-per-week'] > 5).all()
    
    return avg_age_ok and min_hours_ok

# 3. Apply the filter
# This returns a DataFrame containing only the rows from valid groups
filtered_df = df.groupby('occupation').filter(filter_func)

# 4. Show the results
remaining_occupations = filtered_df['occupation'].unique()
print("Occupations that passed the filter:")
print(remaining_occupations)

# Optional: Show the average age of the groups that passed
print("\nStats for passing groups:")
print(filtered_df.groupby('occupation')['age'].mean())
```
