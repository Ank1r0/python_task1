TASK4: Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
import numpy as np

A = np.array([[7, 3, 5], [1, 1, 7], [8, 0, 8],[9,5,3]]) # Sample from your output
B = np.array([[5, 8], [9, 5]])

final_results = []

for row in A:
    matches_B0 = np.intersect1d(row, B[0])
    matches_B1 = np.intersect1d(row, B[1])
    
    if matches_B0.size > 0 and matches_B1.size > 0:
        
        all_matches = np.union1d(matches_B0, matches_B1)
        
        if all_matches.size >= 2:
            final_results.append(row)

result = np.array(final_results)
print("Strict Matching Rows:")
print(result)
```

Задача 4: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.
Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = np.array([2, 2, 2, 3, 3, 3, 5]) 

currElem = None
amount = 0
elem_array = []
step_array = []


for number in x:
    if currElem != number:
        if(amount > 0):
            elem_array.append(currElem)
            step_array.append(amount)

        currElem = number
        amount = 1
    else:
        amount += 1

if(amount > 0):
    elem_array.append(currElem)
    step_array.append(amount)

print(elem_array)
print(step_array)
```
