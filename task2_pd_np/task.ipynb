TASK4: Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
import numpy as np

A = np.array([[7, 3, 5], [1, 1, 7], [8, 0, 8],[9,5,3]]) # Sample from your output
B = np.array([[5, 8], [9, 5]])

final_results = []

for row in A:
    matches_B0 = np.intersect1d(row, B[0])
    matches_B1 = np.intersect1d(row, B[1])
    
    if matches_B0.size > 0 and matches_B1.size > 0:
        
        all_matches = np.union1d(matches_B0, matches_B1)
        
        if all_matches.size >= 2:
            final_results.append(row)

result = np.array(final_results)
print("Strict Matching Rows:")
print(result)
```

Задача 5: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
import numpy as np
from scipy.spatial.distance import cdist
import time

X = np.random.rand(500, 3)
Y = np.random.rand(400, 3)

# --- Method 1: NumPy Broadcasting ---
start_np = time.time()

# We change X from (500, 3) to (500, 1, 3)
# This forces NumPy to compare each X with every Y
dists_np = np.sqrt(np.sum((X[:, np.newaxis, :] - Y[np.newaxis, :, :])**2, axis=2))

end_np = time.time()

# --- Method 2: Scipy cdist ---
start_scipy = time.time()
dists_scipy = cdist(X, Y, metric='euclidean')
end_scipy = time.time()

print(f"NumPy time: {end_np - start_np:.5f}s")
print(f"Scipy time: {end_scipy - start_scipy:.5f}s")

print("Results match:", np.allclose(dists_np, dists_scipy))
```
